### Supplements for histogram: 

__why do these (cumulative) columns have such high frequency in low numerical bins__

The high frequency of low numerical bins in the COVID-19 dataset may be attributed to factors such as the early stages of data collection, the geographic spread of the virus, with regions like Europe, Africa, and America initially experiencing lower case counts due to their distance from the origin of the outbreak. Furthermore, the predominance of Europe, Africa, and America in the dataset, in comparison to other countries, may also contribute to the observed distribution.

__why the distribution of features on such varying scale is not ideal (for modelling) and why there is a need to normalise the features__

The distribution of features on varying scales can obscure the importance of individual features in predictive modeling. When features have vastly different scales, algorithms may assign disproportionate importance to features with larger magnitudes, leading to biased model outcomes.Normalizing features to a common scale (between 0 and 1) ensures that each feature contributes equally to the model's learning process, preventing biases towards certain features based solely on their scale.


### Supplements for pairplot:

__why certain features are or might be highly correlated__

I observed the high correlation between "Confirmed," "Deaths," "Recovered," and "Active" cases. As more individuals are confirmed as COVID-19 positive cases, it is expected that the numbers of deaths, recoveries, and active cases would also increase. This interdependence stems from the fact that confirmed cases serve as the foundation for the subsequent outcomes of recovery, death, or ongoing active cases. Therefore, the correlation reflects the natural flow of the disease process, where confirmed cases lead to outcomes such as recovery or unfortunate fatalities.

__which feature/s would you select to drop and relationship between features__

To prevent multi-collinearity, one approach could involve dropping the "Confirmed" feature. Because "Confirmed" cases are the summation of "Deaths," "Recovered," and "Active" cases. Therefore, including "Confirmed" column may introduce redundancy into the model, potentially inflating the importance of these variables.

It is evident that the cumulative features ('Confirmed,' 'Deaths,' 'Recovered,' and 'Active') exhibit proportional relationships to both cumulative and non-cumulative features. For instance, 'Confirmed' can increase the same amount as 'New cases' have increased. 

On the other hand, the non-cumulative features ('New cases,' 'New deaths,' 'New recovered') do not demonstrate any clear relationships among themselves. Despite fluctuations in daily counts, there appears to be no consistent pattern or correlation between the daily numbers of new cases, deaths, and recoveries
